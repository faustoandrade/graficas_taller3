```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.random.normal(loc=1, scale=1, size=2000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))
```

```python
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.010)
plt.title('SIGMOIDAL')
plt.grid(True)
```
